```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
import sys
path = r'C:\Users\ASUS\Documents\NARIT_internship_data\All_sky_camera_Astropark_Chaingmai\2024-06\2024-06-25\638548759927633225.png'
mask = cv2.imread('mask_delete_5.png',cv2.IMREAD_GRAYSCALE)
def remove_glare(image_path):
    image = cv2.imread(image_path)
    lab = cv2.cvtColor(image, cv2.COLOR_BGR2LAB)
    l,a,b = cv2.split(lab)

    l_eq = cv2.equalizeHist(l)
    lab_eq = cv2.merge((l_eq, a, b))
    lab_eq = cv2.cvtColor(lab_eq,cv2.COLOR_LAB2RGB)
    return np.hstack((image,lab_eq))
rmv = remove_glare(path)
plt.figure(figsize=(20,20))
plt.imshow(rmv)
```
