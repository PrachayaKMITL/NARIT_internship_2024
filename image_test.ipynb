```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
path = r"C:\Users\ASUS\Documents\NARIT_internship_data\All_sky_camera_Astropark_Chaingmai\2024-04\2024-04-04\638477883061862455.png"
def remove_glare(image_path):
    image = cv2.imread(image_path)
    image = cv2.cvtColor(image,cv2.COLOR_BGR2RGBA)
    #lab = cv2.cvtColor(image, cv2.COLOR_RGB2LAB)
    R, G, B, A = cv2.split(image)

    R_eq = cv2.equalizeHist(R)
    A_eq = np.Zero_like(A)
    lab_eq = cv2.merge((R, G, B, A_eq))
    return np.hstack((image,lab_eq))
rmv = remove_glare(path)
plt.figure(figsize=(20,20))
plt.imshow(rmv)
```
