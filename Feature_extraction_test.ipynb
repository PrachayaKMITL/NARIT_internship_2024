```python
import cv2 
from matplotlib import pyplot as plt
```

```python
path = r'C:\Users\User\OneDrive - KMITL\Documents\NARIT_internship_date\638600958679970874.png'
img = cv2.imread(path)
img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.imshow(img,cmap='gray')
```

```python

grad_x = cv2.Sobel(img,cv2.CV_64F,0,1,ksize=11)/100
grad_y = cv2.Sobel(img,cv2.CV_64F,1,0,ksize=11)/100
grad = (grad_x+grad_y)
plt.figure(figsize=(10,10))
plt.imshow(grad,cmap='gray')
```
